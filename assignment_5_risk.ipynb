```python
# Board Game risk
# Author: Andre Machado

import numpy as np 
import matplotlib.pyplot as plt

# Number of battles
num_battles = 1000

# Counters for the wins
attacker_wins = 0
defender_wins = 0

for _ in range(num_battles):
# Generating random numbers to define the dice of attacker
    dice_attacker = np.random.randint(low=1, high=7, size=3)
# Generating random numbers to define the dice of defender
    dice_defender = np.random.randint(low=1, high=7, size=2)
# Calculate maximum and second largest for attacker
    dice_attacker_max = np.max(dice_attacker)
    second_largest_attacker = np.partition(dice_attacker, -2)[-2]   
# Calculate maximum defender
    dice_defender_max = np.max(dice_defender)
# Determine the winner
    if dice_attacker_max > dice_defender_max:
        attacker_wins += 1
    else:
        defender_wins += 1

# Results
print(f"Attacker wins: {attacker_wins}")
print(f"Defender wins: {defender_wins}")

# Calculate and pint the win percentages
total_battles = attacker_wins + defender_wins
attacker_win_percent = attacker_wins / total_battles * 100
defender_win_percent = defender_wins / total_battles * 100
print(f"Attacker win percentage: {attacker_win_percent:.2f}%" )
print(f"Defender win percentage: {defender_win_percent:.2f}%")

# Plotting the results = Pie
labels = ["Attacker win", "Defender win"]
wins = [attacker_wins, defender_wins]
plt.pie(np.array([attacker_win_percent, defender_win_percent]), labels=labels)
plt.show()

# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(labels, wins, color=['blue', 'orange'])
plt.title('Comparison of Wins')
plt.ylabel('Number of Wins')
plt.xlabel('Outcome')
plt.ylim(0, num_battles)
plt.show()
```
